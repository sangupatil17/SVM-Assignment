```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
forest = pd.read_csv("forestfires (1).csv")
```

```python
forest.drop(["month","day"],axis=1,inplace=True)
```

```python
predictors = forest.iloc[:,0:28]
target = forest.iloc[:,28]
```

```python
predictors
```

```python
target
```

```python
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
fires = norm_func(predictors)
```

```python
fires
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size = 0.25, stratify = target)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
```

```python
pred_test_linear = model_linear.predict(x_test)
```

```python
np.mean(pred_test_linear==y_test)
```

```python
acc = accuracy_score(y_test, pred_test_linear) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_test_linear)
```

```python
model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
```

```python
pred_test_poly = model_poly.predict(x_test)
```

```python
np.mean(pred_test_poly==y_test)
```

```python
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
```

```python
pred_test_rbf = model_rbf.predict(x_test)
```

```python
np.mean(pred_test_rbf==y_test) #Accuracy = 76.15%
```
